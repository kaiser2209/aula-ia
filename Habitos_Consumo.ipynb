```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
```

```python
dados = pd.read_csv("dados_brutos.csv")
suportes = apriori(dados, min_support=0.2, use_colnames=True)
suportes #imprime os suportes
```

```python
from mlxtend.frequent_patterns import association_rules
regras = association_rules(suportes, metric="confidence", min_threshold=0.8)
regras
```

```python
regras.to_excel("resultado.xlsx")
```
